```python
import networkx as nx
import random
import pandas as pd
import csv
```

```python
df = pd.read_csv('test3.csv')
df.head
g = nx.from_pandas_edgelist(df, source='source', target='target') 
ListOfNodes = g.nodes()
RandomSample = random.sample(ListOfNodes, 2000)
g.remove_nodes_from(RandomSample)
nx.draw(g)
pos = nx.spring_layout(g, k=10)  # For better example looking
nx.draw(g, pos, with_labels=True, font_size=4)
# labels = {e: g.edges[e]['target'] for e in g.edges}
# nx.draw_networkx_edge_labels(g, pos, edge_labels=labels)
```
